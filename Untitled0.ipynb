```
# Description: This program uses the dual moving average crossoverr to determine when to buy ans sell stock

# import the libraries
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```
# store the data
# col_names = ["Transaction Date/Time","Amount","Value Date","Trans ID","Type","Service Type Details","Payment instrument details","Bank Name","Account","Phone","Name","Transaction Fee","Status"]
ps_stat = pd.read_csv('ps_stat.csv')#, names=col_names
# show the data
ps_stat
```

```
# visualize the data
plt.figure(figsize=(12.5, 4.5))
plt.plot(ps_stat['Amount'], label = 'ps_stat')
plt.title('PS STAT UPDATE')
plt.xlabel('Oct. 1st, 2020 - Oct. 20th, 2020')
plt.ylabel('Amount NGN (#)')
plt.legend(loc='upper left')
plt.show()
```

```
# create the simple moving average with a 5 day window
SMA10 = pd.DataFrame()
SMA10['Amount'] = ps_stat['Amount'].rolling(window=10).mean()
# SMA10
```

```
# create the simple moving average with a 10 days window
SMA20 = pd.DataFrame()
SMA20['Amount'] = ps_stat['Amount'].rolling(window=20).mean()
# SMA20
```

```
# visualize the data
plt.figure(figsize=(12.5, 4.5))
plt.plot(ps_stat['Amount'], label = 'ps_stat')
plt.plot(SMA10['Amount'], label = 'SMA10')
plt.plot(SMA20['Amount'], label = 'SMA20')
plt.title('PS STAT UPDATE')
plt.xlabel('Oct. 1st, 2020 - Oct. 20th, 2020')
plt.ylabel('Amount NGN (#)')
plt.legend(loc='upper left')
plt.show()
```

```
# create a new data frame to store all the data 
data = pd.DataFrame()
data['ps_stat'] = ps_stat['Amount']
data['SMA10'] = SMA10['Amount']
data['SMA20'] = SMA20['Amount']
# data
```

```
# create a function to signal when to buy and sell the asset/stock
def buy_sell(data):
  sigPriceBuy = []
  sigPriceSell = []
  flag = -1

  for i in range(len(data)):
    if data['SMA10'][i] > data['SMA20'][i]:
      if flag != 1:
        sigPriceBuy.append(data['ps_stat'][i])
        sigPriceSell.append(np.nan)
        flag = 1

      else:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(np.nan)
    elif data['SMA10'][i] < data ['SMA20'][i]:
      if flag != 0:
        sigPriceBuy.append(np.nan)
        sigPriceSell.append(data['ps_stat'][i])
        flag = 0
      else:
        sigPriceBuy.append(np.nan)
        sigPriceBuy.append(np.nan)
    else:
      sigPriceBuy.append(np.nan)
      sigPriceBuy.append(np.nan)
    
  return (sigPriceBuy, sigPriceSell)
```

```
# store the buy and sell data into variable
# echoing wrong data look into the loop 
buy_sell = buy_sell(data)
for buy_sell_value in buy_sell:
  data['Buy_Signal_Price'] = buy_sell_value[0]
  data['Sell_Signal_Price'] = buy_sell_value[1]
```

```
# show the data
data
```

```
# visualize the data and the strategy to buy and sell the stock
plt.figure(figsize=(12.6, 4.6))
plt.plot(data['ps_stat'], label = 'PS STAT', alpha = 0.35)
plt.plot(data['SMA10'], label = 'SMA10', alpha = 0.35)
plt.plot(data['SMA20'], label = 'SMA20', alpha = 0.35)
plt.scatter(data.index, data['Buy_Signal_Price'], label = 'Buy', marker = '^', color='green')
plt.scatter(data.index, data['Sell_Signal_Price'], label = 'Sell', marker = '^', color='red')
plt.title('PS STATISTICS STOCK-SIGNALS')
plt.xlabel('Oct. 1st, 2020 - Oct. 30th, 2020')
plt.ylabel('Amount NGN (#)')
plt.legend(loc='upper left')
plt.show()
```
